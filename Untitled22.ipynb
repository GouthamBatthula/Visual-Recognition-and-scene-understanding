```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0, 4*np.pi, 1000)
y=np.sin(x)+0.3*np.sin(3*x)+0.2*np.cos(7*x)
plt.figure()
plt.plot(x, y);
plt.xlabel("x")
plt.ylabel("Y")
plt.title("continuous signal")
plt.grid(True)
plt.show()
```

```python
sampling_factor = 50
x_sampling = x[::sampling_factor]
y_sampling = y[::sampling_factor]
plt.figure()
plt.plot(x_sampling,y_sampling)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Sampling signal")
plt.grid(True)
plt.show()
```

```python
from matplotlib.artist import setp
#quantization reducting amplitude resoultion
num_levels = 8
y_min, y_max = y.min(), y.max()
step = (y_max - y_min) / num_levels
y_quantized = np.round((y - y_min) / step) * step + y_min

#plotting the signal with sampling and quantization
plt.figure(figsize=(10,4))
plt.plot(x,y, label='Continuous Signal', alpha=0.75, color='Blue')
markerline, stemlines, baselines = plt.stem(x_sampling, y_sampling, linefmt='r-', markerfmt='ro', basefmt=' ', label='Sampled Signal')
plt.setp(markerline, alpha=0.2)
plt.setp(stemlines, alpha=0.2)
#Horizontal quantization lines
for i in range(num_levels + 1):
  y_line = y_min + i * step
  plt.axhline(y_line, color='gray', linestyle='--', linewidth=0.5, alpha=0.6)

plt.plot(x, y_quantized, color='green', linestyle=':', label='Quantized Signal')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sampling and Quantization of a Signal')
plt.legend()
plt.grid(True)
plt.show()
```
